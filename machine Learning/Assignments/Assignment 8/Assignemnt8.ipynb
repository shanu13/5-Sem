```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.utils import shuffle
from sklearn.metrics import accuracy_score
import random
```

```python
def predict(xtest,ytest,weight):
  ypred = np.ones(ytest.shape[0])
  hypo = np.dot(xtest,weight)
  print(hypo)

  for i in range(0,hypo.shape[0]-1):
    if(hypo[i] >=0.5):
      ypred[i] = 1
    if(hypo[i] <0.5):
      ypred[i] = 0

  test_acc = accuracy_score(ytest,ypred)
  return test_acc


def add_intercept(xtrain):
  intercept = np.ones((xtrain.shape[0],1))
  # print(intercept)
  # print(np.concatenate((intercept,xtrain),axis=1))
  return np.concatenate((intercept,xtrain),axis=1);

def perceptron(xtrain,ytrain,same_weight):
  xtrain = add_intercept(xtrain)
  used_index = []
  weight = np.zeros(xtrain.shape[1])

  for i in range(0,xtrain.shape[1]):
    weight[i] = 0.1
  index = random.randint(0,xtrain.shape[0]-1)
  xtrain,ytrain = shuffle(xtrain,ytrain)
  used_index.append(index)
  epoch = 1
  converged = False
  while not converged:
    print("epoch",epoch)
    hypothesis = np.dot(xtrain[index],weight)
    if(ytrain[index] > 0 and hypothesis < 0):
      weight = weight + xtrain[index]

    if(ytrain[index] < 0 and hypothesis >0):
      weight = weight - xtrain[index]

    index = random.randint(0,xtrain.shape[0]-1)
    used_index.append(index)

    if(len(set(used_index)) == xtrain.shape[0]):
      # check accuracy
      score = predict(xtrain,ytrain,weight)
      if(score*100 == 100):
        print("succesfully converged")
        converged = True
      epoch = epoch+1
  
  return weight
    

     

  
```

```python
x_train = np.array([[0,1],[1,1],[0,0],[1,0]])
print(x_train)
y_train = np.array([1, 1, 0, 1])
print(y_train)
weight = perceptron(x_train,y_train,same_weight= False)
print(weight)

```
