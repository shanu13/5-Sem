```python
data = pd.DataFrame(dataset)
for r in range(10,100,10):
  x = data.sample(frac=r/100)
  y = data.drop(x.index)
  print(f"X:Y ratio is {r/100, 1-r/100}")
  print("\n",x)
  print("\n",y)
  print("\n")
```

#Write a program to plot a graph (using matplotlib) for the

```python
import matplotlib.pyplot as plt
```

```python
dataset2 = pd.read_csv('student_scores.csv')
data2 = pd.DataFrame(dataset2)
```

```python
xdata2 = data2.iloc[:,:-1]
ydata2 = data2.iloc[:,-1]
xdata2 = np.array(xdata2)
ydata2 = np.array(ydata2) 
```

```python
print(xdata2)
print(ydata2)
```

```python
plt.scatter(xdata2,ydata2, color="green")
plt.xlabel("Hours")
plt.ylabel("Scores")
plt.title("student scores  as per study hours")
plt.show()
```
